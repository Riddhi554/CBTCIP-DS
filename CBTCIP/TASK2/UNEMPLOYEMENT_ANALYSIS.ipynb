```python
import pandas as pd
data = pd.read_csv("Unemployment in India - Unemployment in India.csv")
print(data.head())
```

```python
print(data.columns)

print(data.isnull().sum())

print(data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=data)
plt.title(' Unemployment Rate Over Time')
plt.xlabel(' Date')
plt.ylabel(' Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
start_date = '01-03-2020'
end_date = '30-06-2020'
covid_period_data = data[(data['Date'] >= start_date) & (data['Date'] <= end_date)]
average_unemployment_rate_covid = covid_period_data['Estimated Unemployment Rate (%)'].mean()
print(f'Average Unemployment Rate during Covid-19: {average_unemployment_rate_covid:.2f}')
```
